```python
##Prime number generator within range(0 to num of your choice)
def prime_gen(rng):
    primearr=[2]
    for x in range(3,rng):
        for y in range(2, x):
            if(x % y == 0):
                break
        else:
            prime = x
            primearr.append(prime)
    else:
        pass
             
    return primearr
rng = 50000
primes = prime_gen(rng)
##print(primes)
```
